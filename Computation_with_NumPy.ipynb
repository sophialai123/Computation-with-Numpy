NumPy's most amazing feature is the **powerful** ndarray.

#### 1-Dimensional Arrays (Vectors)

```python
#create a 1-dimensional array
my_array = np.array([1.1, 9.2, 8.1, 4.7])
```

```python
#see my_array is 1 dimensional by looking at its shape,show rows and columns
my_array.shape
```

```python
#access an element in a ndarray by that element's index(same as python list)
my_array[2]
```

```python
#check how many dimensions use the ndim attribute: 
my_array.ndim
```

#### 2-Dimensional Arrays (Matrices)

```python
#create a 2-dimensional array
array_2d = np.array([[1, 2, 3, 9], 
                     [5, 6, 7, 8]])
```

```python
#This array has 2 rows and 4 columns.
#NumPy refers to the dimensions as axes, 
#so the first axis has length 2 and the second axis has length 4.
array_2d.shape
```

```python
print(f'array_2d has {array_2d.ndim} dimensions')
print(f'It has {array_2d.shape[0]} rows and {array_2d.shape[1]} columns')
print(array_2d)
```

```python
#To access a particular value, provide an index for each dimension. 
#We have two dimensions, so we need to provide an index for the row and for the column. 

#Here’s how to access the 3rd value in the 2nd row
array_2d[1,2]
```

```python
#use the : operator access all the values in the first row
array_2d[0,:]
```

# Linear Algebra with Vectors

```python
v1 = np.array([4, 5, 2, 7])
v2 = np.array([2, 1, 3, 3])

#And you add them together, the numbers will add up
v1 + v2
```

```python
#Multiplying the two vectors together, since 4x2=8, 5x1=5 and so
#Python List, this operation would not work at all
print(v1 * v2)
```

```python
# Python Lists vs ndarrays
list1 = [4, 5, 2, 7]
list2 = [2, 1, 3, 3]

#adding them together would just concatenate the lists.
list1 + list2
```

# Broadcasting and Scalars


```python
#sort of operation between an array and a single number. In mathematics, this single number is often called a scalar.
#NumPy will make the shape of the smaller array - our scalar - compatible with the larger array. 
#This is what the documentation refers to when it mentions the term "broadcasting".
array_2d = np.array([[1, 2, 3, 4], 
                     [5, 6, 7, 8]])

#The scalar operates on an element by element basis.
array_2d + 10
```

```python
# broadcasting
array_2d * 5
```

# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

**Challenge**: Let's multiply `a1` with `b1`. Looking at the wikipedia example above, work out the values for c12 and c33 on paper. Then use the [`.matmul()`](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html) function or the `@` operator to check your work. 

```python
#use the .matmul() function 
c = np.matmul(a1, b1)
print(f'Matrix has {c.shape[0]} rows and {c.shape[1]} columns.')
c
```

```python
#or the @ operator to check your work.
a1 @ b1
```

```python
"""But how did the calculations arrive at 25 for c12 and 28 for c33? Substituting the number into the formula we get:
c12 = 1*1 + 3*8 = 1 + 24 = 25

c33 = 6*3 + 2*5 = 18 + 10 = 28"""
```

# Manipulating Images as ndarrays


```python
#The Scipy library contains an image of a racoon under 'miscellaneous' (misc). 
img = misc.face()

#display it using Matplotlib's .imshow()
plt.imshow(img)
```

**Challenge**: Convert the image to black and white. The values in our `img` range from 0 to 255. 
* Divide all the values by 255 to convert them to sRGB, where all the values are between 0 and 1. 
* Next, multiply the sRGB array by the `grey_vals` to convert the image to grey scale. 
* Finally use Matplotlib's [`.imshow()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html) together with the colormap parameter set to gray `cmap=gray` to look at the results. 

```python
#Divide all the values by 255 to convert them to sRGB,
#where all the values are between 0 and 1
sRGB_array = img / 255
```

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
#multiply them together with the @ operator
img_gray = sRGB_array @ grey_vals

# or the .matmul() function
#img_gray = np.matmul(sRGB_array, grey_vals)
```

```python
#The cmap parameter is important here. If we leave it out the function will not know that is dealing with a black and white image.
plt.imshow(img_gray,cmap='gray')
```

#### Challenge Solutions

```python
#flip the grayscale image upside down
img_gray = np.flip(img_gray)
plt.imshow(img_gray,cmap='gray')
```

```python
#To rotate the image, np.rot90()
img = np.rot90(img)
plt.imshow(img)
```

```python
#Invert (i.e., solarize) the colour image. 
solar_img = 255 - img
plt.imshow(solar_img)
```
